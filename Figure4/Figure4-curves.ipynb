```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

plt.rcParams['figure.figsize'] = 21, 15
plt.rcParams['xtick.labelsize']=65
plt.rcParams['ytick.labelsize']=65
plt.rcParams['ytick.right'] = plt.rcParams['ytick.labelright'] = False
plt.rcParams['ytick.left'] = plt.rcParams['ytick.labelleft'] = True
plt.rcParams.update({'figure.autolayout': False})
plt.rcParams['axes.linewidth'] = 5
plt.rcParams['font.family'] = 'Arial'
```

```python
file_path = "Figure4.xlsx"
df = pd.read_excel(file_path)

time = df["time"] / 3600

plot_config = [
    {
        "cols": ["gRNA 14", "gRNA 14.1", "gRNA 14.2"],
        "label": "gRNA 14",
        "color": "#ff8300",
    },
    {
        "cols": ["gRNA 14 DNA", "gRNA 14 DNA.1", "gRNA 14 DNA.2"],
        "label": "gRNA 14 + dsDNA",
        "color": "#ffb100",
    },
]

fig, ax = plt.subplots()

for cfg in plot_config:
    for col in cfg["cols"]:
        ax.plot(
            time,
            df[col],
            linewidth=15,
            alpha=0.3,
            color=cfg["color"],
            zorder=1
        )
    mean_values = df[cfg["cols"]].mean(axis=1)
    ax.plot(
        time,
        mean_values,
        label=cfg["label"],
        color=cfg["color"],
        linewidth=6,
        linestyle='--',
        zorder=3
    )

custom_legend = [
    Line2D([0], [0], color=cfg["color"], lw=12, label=cfg["label"])
    for cfg in plot_config
]

plt.title("Background DNA N200 at 0.2 µM", fontsize=60, weight='bold', pad=15)
ax.set_xlim([-0.1, 7])
ax.set_ylim([-10, 220])
ax.tick_params(axis='both', which='major', length=15, width=6)

ax.set_xlabel('Time (hours)', fontsize=65)
ax.set_ylabel('Intensity (n.u.)', fontsize=65)

ax.legend(
    handles=custom_legend,
    loc='upper left',
    bbox_to_anchor=(0.35, 0.35),
    fontsize=45,
    frameon=False
)

plt.subplots_adjust(right=0.75)
plt.savefig("Figure4-gRNA14-kinetics.svg", format="svg", dpi=300)
plt.show()
```
